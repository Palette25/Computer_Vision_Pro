```python
import cv2
```

```python
def detectFace_Video(file):
    dst_name = 'videos/output.avi'

    # Use cascade classifier
    face_classifier = cv2.CascadeClassifier('pretrained_classifier.xml')
    # Read the video
    cap = cv2.VideoCapture(file)
    # Define output video writer
    outputWriter = cv2.VideoWriter(dst_name, cv2.VideoWriter_fourcc('I', '4', '2', '0'), 
                                   cap.get(5), (160, 120))
    
    # Detecting faces frame by frame
    retval, frame = cap.read()
    index = 0
    
    while retval:
        print('Processing %dth frame...' % index)
        index += 1
        
        out_frame = frame
        # Turn it into gray-image
        gray_img = cv2.cvtColor(out_frame, cv2.COLOR_BGR2GRAY)
        # Detect face
        faces = face_classifier.detectMultiScale(gray_img, 1.3, 5)
        # Draw face bounding boxes
        for (x, y, w, h) in faces:
            out_frame = cv2.rectangle(out_frame, (x, y), (x+w, y+h), (255, 255, 0), 2)
        outputWriter.write(out_frame)
        
        retval, frame = cap.read()

    cap.release()
    # Show video
    showOutput(dst_name)

    return dst_name
```

```python
def showOutput(dst):
    # Read result detected video
    cap = cv2.VideoCapture(dst)
    # Loop to show video frames
    while cap.isOpened():
        retval, frame = cap.read()
        if not retval:
            print('Output video end~')
            # Destroy windows
            cv2.destroyAllWindows()
            cap.release()
            return
        cv2.imshow('Video', frame)
        cv2.waitKey(50)
```

```python
if __name__ == '__main__':
    filename = 'videos/input.avi'
    dst = detectFace_Video(filename)
```
